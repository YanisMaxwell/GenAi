```python
nombre = int(input(" Donne un nombre : "))

def check_even_or_odd(number):
    
    return "Even" if number % 2 == 0 else "Odd"

resultat = check_even_or_odd(nombre)
print(resultat)
```

Exercice 4

```python
total_1 = [1,2,3]

def sum_list(numbers):
    
    total = 0
    for i in numbers:
         total += i

         return total

bonne_reponse = sum_list(total_1)
print(bonne_reponse)
```
